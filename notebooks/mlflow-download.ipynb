```python
from mlflow.tracking import MlflowClient

import pandas as pd

from tqdm import tqdm
```

```python
# Initialize MLflow Client
client = MlflowClient(tracking_uri="http://potato.felk.cvut.cz:2222")

experiment_names = [
    "pelesjak_linear_sage_pretrain",
    "pelesjak_resnet_sage_pretrain",
    "pelesjak_linear_sage_nopretrain",
    "pelesjak_resnet_sage_nopretrain",
]

for experiment_name in experiment_names:
    experiment = client.get_experiment_by_name(experiment_name)
    if experiment is None:
        raise ValueError(f"Experiment '{experiment_name}' not found.")
    experiment_id = experiment.experiment_id

    # Get all runs for the experiment
    runs = client.search_runs(experiment_ids=[experiment_id], max_results=10000)

    runs_dict = []
    for r in tqdm(runs.to_list()):
        r_info = r.info.__dict__
        r_data = {k: v for d in r.data.to_dictionary().values() for k, v in d.items()}

        # metrics, heigher_is_better, tune_metric = get_metrics(
        #     r_data["dataset"], r_data["task"]
        # )
        # tune_history = client.get_metric_history(r.info.run_id, f"val_{tune_metric}")
        # best_step = get_best_step(tune_history, heigher_is_better)

        # for m in metrics:
        #     m_val_hist = client.get_metric_history(r.info.run_id, f"val_{m}")
        #     r_data[f"best_val_{m}"] = get_step_value(m_val_hist, best_step)
        #     r_data[f"val_hist_{m}"] = [m.to_dictionary() for m in m_val_hist]
        #     m_test_hist = client.get_metric_history(r.info.run_id, f"test_{m}")
        #     r_data[f"best_test_{m}"] = get_step_value(m_val_hist, best_step)
        #     r_data[f"test_hist_{m}"] = [m.to_dictionary() for m in m_test_hist]

        runs_dict.append({**r_info, **r_data})

    df = pd.DataFrame(runs_dict)

    df.to_csv(f"./{experiment_name}.csv")
```
